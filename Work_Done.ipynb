```python
#Insight 1: Top 5 parties with the most wins
top_5_parties = election_results_df.nlargest(5, 'Won')
print(top_5_parties)
```

```python
#Insight 2: Average number of wins per party
average_wins = election_results_df['Won'].mean()
print(f"Average number of wins per party: {average_wins:.2f}")
```

```python
#Insight 3: Parties with only 1 win
single_win_parties = election_results_df[election_results_df['Won'] == 1]
print(single_win_parties)
```

```python
#Insight 4: Top 3 parties with the most leads
top_3_leads = election_results_df.nlargest(3, 'Leading')
print(top_3_leads)   
```

```python
#Insight 5: Parties with no leads
no_leads_parties = election_results_df[election_results_df['Leading'] == 0]
print(no_leads_parties)
```

```python
#Graph Insight 1: Bar chart of the top 5 parties with the most wins
import matplotlib.pyplot as plt

top_5_parties = election_results_df.nlargest(5, 'Won')
plt.bar(top_5_parties['Party'], top_5_parties['Won'])
plt.xlabel('Party')
plt.ylabel('Wins')
plt.title('Top 5 Parties with the Most Wins')
plt.show()
```

```python
#Graph Insight 2: Pie chart of the number of parties with different number of wins
import matplotlib.pyplot as plt

# Create a dictionary of the number of parties with different number of wins
win_counts = {}
for i in range(1, 11):
    win_counts[i] = len(election_results_df[election_results_df['Won'] == i])

# Create a pie chart
plt.pie(win_counts.values(), labels=win_counts.keys(), autopct='%1.1f%%')
plt.title('Number of Parties with Different Number of Wins')
plt.show()
```

```python
#Graph Insight 3: Box plot of the number of wins per party
import matplotlib.pyplot as plt

# Create a box plot of the number of wins per party
plt.boxplot(election_results_df['Won'].astype(int))
plt.xlabel('Party')
plt.ylabel('Wins')
plt.title('Number of Wins Per Party')
plt.show()
```

```python
#Graph Insight 4: Bar chart of the top 3 parties with the most leads
import matplotlib.pyplot as plt

top_3_leads = election_results_df.nlargest(3, 'Leading')
plt.bar(top_3_leads['Party'], top_3_leads['Leading'])
plt.xlabel('Party')
plt.ylabel('Leads')
plt.title('Top 3 Parties with the Most Leads')
plt.show()
```
